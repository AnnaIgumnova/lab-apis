```python
artists = ["Florence + The Machine", "Rosalía", "Lou Doillon"]
```

```python
def get_artist_id(artist_name):
    # Search Spotify for this artist
    result = sp.search(q=artist_name, type="artist", limit=1)
    
    items = result["artists"]["items"]  # list of matching artists
    
    if not items:          # if the list is empty, artist not found
        return None
    
    artist_id = items[0]["id"]  # take the first artist's id
    return artist_id
```

```python
print(get_artist_id("Florence + The Machine"))
print(get_artist_id("Rosalia"))
print(get_artist_id("Lou Doillon"))
```

```python
def get_top_tracks(artist_name):
    # 5 topt tracks of the artist

    artist_id = get_artist_id(artist_name)
    
    if artist_id is None:
        # Artist not found, return empty list
        return []
    
    # Get the top tracks for this artist
    top_tracks_data = sp.artist_top_tracks(artist_id)
    
    # top_tracks_data["tracks"] is a list of track dictionaries
    tracks = top_tracks_data["tracks"]
    
    # Take only first 5 track names
    top_track_names = [track["name"] for track in tracks[:5]]
    
    return top_track_names
get_top_tracks("Rosalia")
```

```python
top_tracks_list = []

for artist in artists:
    tracks_for_artist = get_top_tracks(artist)
    top_tracks_list.append(tracks_for_artist)

top_tracks_list
```

```python
from spotipy.exceptions import SpotifyException
```

```python
def find_related_artists(artist_name):
    """
    Takes an artist name (string),
    returns a list with the names of the first 5 related artists.

    If Spotify returns an error (like 404 Not Found),
    we just print a message and return an empty list instead of crashing.
    """
    artist_id = get_artist_id(artist_name)
    
    if artist_id is None:
        print(f"No artist found for '{artist_name}'.")
        return []
    
    try:
        # Try to call Spotify API
        related_data = sp.artist_related_artists(artist_id)
    except SpotifyException as e:
        # If Spotify gives an error (like 404), handle it nicely
        print(f"Could not get related artists for '{artist_name}': {e}")
        return []
    except Exception as e:
        # Any other unexpected error
        print(f"Something went wrong for '{artist_name}': {e}")
        return []
    
    # If we reach here, the call worked and we have data
    related_artists = related_data["artists"]
    related_artist_names = [artist["name"] for artist in related_artists[:5]]
    
    return related_artist_names


find_related_artists("Taylor Swift")
```

```python
print(find_related_artists("Florence + The Machine"))
print(find_related_artists("Rosalía"))
print(find_related_artists("Lou Doillon"))
```

## Playlists

The `sp.featured_playlists()` method in `spotipy` fetches a list of Spotify's featured playlists at a given moment. These are curated playlists that Spotify often highlights on the platform's homepage. The method provides a snapshot of the playlists that are being promoted or featured by Spotify at the time of the request.

Once you've fetched the featured playlists, you can extract their IDs (and other details).

```python
def get_example_playlists(query="hits", limit=20):
    """
    Search Spotify for playlists matching a query word 
    (e.g. 'hits', 'chill', 'pop') and return a list of playlist objects.
    """
    result = sp.search(q=query, type="playlist", limit=limit)
    playlists = result["playlists"]["items"]
    return playlists
```

```python
playlists = get_example_playlists(query="hits", limit=10)

print("Type of playlists:", type(playlists))
print("Number of playlists:", len(playlists))

# Show first 5 playlists safely
for i, p in enumerate(playlists[:5], start=1):
    if p is None:
        print(f"{i}. Skipping: playlist is None")
        continue
    
    print(f"{i}. Name: {p.get('name')}")
    print("   ID:  ", p.get("id"))
    print("-" * 40)
```

```python
chosen_playlist_id = playlists[1]["id"]   # or any index you like
chosen_playlist_id
```

```python
playlist_details = sp.playlist(chosen_playlist_id)

playlist_name = playlist_details["name"]
playlist_description = playlist_details["description"]
total_tracks = playlist_details["tracks"]["total"]

print("Name:       ", playlist_name)
print("Description:", playlist_description)
print("Total tracks:", total_tracks)
```

```python
playlist_items = sp.playlist_items(chosen_playlist_id, limit=10)
items = playlist_items["items"]

print("First 10 tracks:")
for i, item in enumerate(items, start=1):
    track = item["track"]
    track_name = track["name"]
    print(f"{i}. {track_name}")
```

```python
track_artists_dict = {}

for item in items:
    track = item["track"]
    track_name = track["name"]
    
    artist_names = [artist["name"] for artist in track["artists"]]
    track_artists_dict[track_name] = artist_names

for track_name, artist_names in track_artists_dict.items():
    print(f"Track: {track_name}")
    print("Artists:", ", ".join(artist_names))
    print("-" * 40)
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

## Exercise 2: Unraveling the World of Playlists


1. **Featured Exploration**: 
   - Fetch the list of Spotify's current featured playlists. 
   - Extract and display the names and IDs of the top 5 featured playlists.
   
2. **Deep Dive**:
   - Choose any one of the top 5 featured playlists (you can choose the one you personally find most interesting or simply pick one randomly).
   - Fetch and display its name, description, and total track count.

3. **Track-tastic**:
   - Extract and display the names of the first 10 tracks in the chosen playlist.

4. **Artistic Flair**:
   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python

playlists = get_example_playlists(query="hits", limit=10)

print("Top 5 playlists (name + ID):\n")

valid_playlists = []

for i, pl in enumerate(playlists[:5], start=1):
    if pl is None:
        print(f"{i}. Skipping: playlist is None\n")
        continue
    
    print(f"{i}. Name: {pl.get('name', 'Unknown')}")
    print(f"   ID:   {pl.get('id', 'Unknown')}\n")
    valid_playlists.append(pl)

# Make sure we have at least one good playlist
if not valid_playlists:
    raise ValueError("No valid playlists found. Try a different query word!")
```

```python
chosen_playlist = valid_playlists[1]
chosen_id = chosen_playlist["id"]

playlist_details = sp.playlist(chosen_id)

print("\n=== Deep Dive into Chosen Playlist ===")
print("Name:        ", playlist_details.get("name"))
print("Description: ", playlist_details.get("description"))
print("Total tracks:", playlist_details["tracks"]["total"])
```

```python
tracks_response = sp.playlist_tracks(chosen_id)
first_10_tracks = tracks_response["items"][:10]

print("\nFirst 10 tracks:\n")
for i, item in enumerate(first_10_tracks, start=1):
    track = item.get("track")
    if track is None:
        print(f"{i}. Skipping: track is None")
        continue
    print(f"{i}. {track.get('name')}")
```

```python

track_artists = {}

for item in first_10_tracks:
    track = item.get("track")
    if track is None:
        continue
        
    track_name = track.get("name")
    artist_names = [artist["name"] for artist in track.get("artists", [])]
    track_artists[track_name] = artist_names

print("\nDictionary of tracks and their artists:\n")
for track_name, artists in track_artists.items():
    print(f"- {track_name}: {artists}")
```
